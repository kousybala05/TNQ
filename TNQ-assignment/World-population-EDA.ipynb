```python
df.rename(columns = {'2022 Population':'2022_Population','2020 Population':'2020_Population', '2015 Population':'2015_Population'
                    ,'2010 Population':'2010_Population','2000 Population':'2000_Population','1990 Population':'1990_Population','1980 Population':'1980_Population', '1970 Population':'1970_Population'
                    ,'Area (km²)':'Area(km²)','Density (per km²)':'Density(per km²)','Growth Rate':'Growth_Rate',
                    'World Population Percentage':'World_Population_Percentage'}, inplace = True)
```

```python
numeric_features=df.select_dtypes(include=[np.number])
print(numeric_features.columns)
```

```python
cat_features=df.select_dtypes(include=[np.object])
print("Categorical colum names-->")
print(cat_features.columns)
```

```python
def boxplot(column):
    sns.boxplot(data=df,x=df[f"{column}"])
    plt.title(f"Boxplot of Population dataset {column}")
    plt.show()
```

```python
boxplot('Rank')
boxplot('World_Population_Percentage')
boxplot('Area(km²)')
boxplot('Density(per km²)')
boxplot('2022_Population')
boxplot('Growth_Rate')
boxplot('1990_Population')
```

```python
print(df['World_Population_Percentage'].quantile(0.10))
print(df['World_Population_Percentage'].quantile(0.90))
```

```python
df["World_Population_Percentage"] = np.where(df["World_Population_Percentage"] <0.0, 0.0,df['World_Population_Percentage'])
df["World_Population_Percentage"] = np.where(df["World_Population_Percentage"] >0.747, 0.747,df['World_Population_Percentage'])
```

```python
print(df['Area(km²)'].quantile(0.10))
print(df['Area(km²)'].quantile(0.90))
```

```python
df["Area(km²)"] = np.where(df["Area(km²)"] <247.4, 247.4,df['Area(km²)'])
df["Area(km²)"] = np.where(df["Area(km²)"] >1234445.5000000002, 1234445.5000000002,df['Area(km²)'])
```

```python
print(df['Density(per km²)'].quantile(0.10))
print(df['Density(per km²)'].quantile(0.90))
```

```python
df["Density(per km²)"] = np.where(df["Density(per km²)"] <15.837280000000002, 15.837280000000002,df['Density(per km²)'])
df["Density(per km²)"] = np.where(df["Density(per km²)"] >522.8226, 522.8226,df['Density(per km²)'])
```

```python
print(df['2022_Population'].quantile(0.10))
print(df['2022_Population'].quantile(0.90))
```

```python
df["2022_Population"] = np.where(df["2022_Population"] <48225.200000000004, 48225.200000000004,df['2022_Population'])
df["2022_Population"] = np.where(df["2022_Population"] >59636961.70000002, 59636961.70000002,df['2022_Population'])
```

```python
print(df['1990_Population'].quantile(0.10))
print(df['1990_Population'].quantile(0.90))
```

```python
df["1990_Population"] = np.where(df["1990_Population"] <46476.6, 46476.6,df['1990_Population'])
df["1990_Population"] = np.where(df["1990_Population"] >42913893.20000007, 42913893.20000007,df['1990_Population'])
```
